3. Write a python program to access the search bar and search button on images.google.com and scrape 10
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’. 

```python
#Importing required libraries

import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException, ElementNotInteractableException, TimeoutException, InvalidArgumentException
from selenium.webdriver.chrome.service import Service

import time
import os
import requests

# Opening the chrome browser
driver = webdriver.Chrome()

# opening Google Images home page on browser
driver.get('https://images.google.com')
time.sleep(2)

#maximizing the browser window
driver.maximize_window()

# Keywords to search
keywords = ['fruits','cars','Machine Learning', 'Guitar', 'Cakes']
image_count=10
image_urls=[]

# Create a folder to save images
save_dir = r"C:\Users\PFaizaNahyan\Documents\FlipRobo"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

#locate the search bar element to input given keywords in loop
for keyword in keywords:
    search_bar=driver.find_element(By.XPATH,'//textarea[@class="gLFyf"]')
    search_bar.clear() # to clear previous input
    
    search_bar.send_keys(keyword)
    
    search_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span').click()
    time.sleep(3)

    # Wait for images to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[@class="H8Rx8c"]/g-img/img')))
        
    # to collect image URLs
    images=driver.find_elements(By.XPATH,'//div[@class="H8Rx8c"]/g-img/img')
    
    for image in images[:image_count]:
        image_url=image.get_attribute('src') or image.get_attribute('data-src') or image.get_attribute('srcset')
        if image_url and image_url.startswith('http'):
            image_urls.append(image_url)

    # going back to Google images home page
    driver.back()
    time.sleep(2)

# For downloading the images
for i, image_url in enumerate(image_urls):
    if i >= image_count:
        break
    print(f"Downloading {i+1} of {image_count} images")
    try:
        response=requests.get(image_url)
        if response.status_code ==200:
            with open(os.path.join(save_dir, f'image_{i}.jpg'), 'wb') as file:
                file.write(response.content)
        else:
            print(f"Failed to download {image_url}: Status code {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {image_url}: {e}")
            
    
# Close the browser
driver.quit()
```

7Q.Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped:
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”. 

```python
driver = webdriver.Chrome()
driver.get("https://www.forbes.com/billionaires/")

driver.maximize_window()
time.sleep(10)

# Extracting from the page:

details=[]
a=driver.find_elements(By.XPATH,'//div[@class="Table_tableRow__lF_cY"]')
for i in a:
    b=i.text
    details.append(b)
    
time.sleep(4)
```

```python
details
```

```python
import re
# Define a regex pattern to capture each part (Rank, Name, Net Worth, Age, Country, Source, Industry)
pattern = r'(\d+\.)\n([^\n]+)\n(\$[^\n]+)\n(\d+)\n([^\n]+)\n([^\n]+)\n([^\n]+)'

# Parse each entry using regex
parsed_data = []
for entry in details:
    match = re.findall(pattern, entry)
    if match:
        parsed_data.append(match[0])  # Add the match to the parsed_data list
    else:
        print(f"Skipping entry due to incorrect format: {entry}")
        
billionaires=pd.DataFrame(parsed_data,columns=['Rank','Name','Net Worth','Age','Country/Territory','Source','Industry'])
billionaires
```

```python
# Save the data to a CSV file
billionaires.to_csv('forbes_billionaires.csv', index=False)

# Close the browser
driver.quit()

print("Data scraping completed and saved to 'forbes_billionaires.csv'.")
```

8Q.Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
from any YouTube Video. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up the WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run headless Chrome (optional)
driver = webdriver.Chrome()

# Open the YouTube video
driver.get('https://www.youtube.com/watch?v=5cx7rvMvAWo')

# Wait for comments to load and scroll to load more comments
scroll_pause_time = 2  # Adjust the pause time as needed
scrolls = 10  # Adjust the number of scrolls as needed

for _ in range(scrolls):
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(scroll_pause_time)

# Extract comments, upvotes, and times
# Updated selectors based on YouTube's comment structure
try:
    comments = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.ID,'content-text')))
    upvotes = driver.find_elements(By.ID,'vote-count-middle')
    times = driver.find_elements(By.ID,'published-time-text')

    # Ensure lists are the same length
    min_len = min(len(comments), len(upvotes), len(times))

    # Store the extracted data
    extracted_data = []
    for i in range(min_len):
        extracted_data.append({
            'comment': comments[i].text,
            'upvote': upvotes[i].text if i < len(upvotes) else '-',
            'time': times[i].text if i < len(times) else '-'
        })
except Exception as e:
    print("Error extracting data: ", e)

# Close the WebDriver
driver.quit()

# Print the extracted data
for data in extracted_data:
    print(data)
```

reviews, privates from price, dorms from price, facilities and property description. 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time

def scrape_hostelworld_london():
    # Setup Selenium with Chrome options
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Run headless Chrome
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")

    driver = webdriver.Chrome(options=chrome_options)
    
    # URL of the Hostelworld homepage
    base_url = "https://www.hostelworld.com/"
    driver.get(base_url)
    
    # Close any popups or survey if they appear
    try:
        WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '/html/body/div[6]/div/div[2]/div[1]/div[1]/div[2]/button'))).click()
    except Exception as e:
        print("No popup found or unable to close the popup: ", e)

    
    # Locate the search input box and ensure it is clickable
    try:
        search_box = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, "native-input"))
        )
        search_box.click()
        search_box.send_keys("London")
        time.sleep(2)  # Wait for autocomplete suggestions to load
        
        # Handle autocomplete suggestions
        try:
            suggestions = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "/html/body/div[3]/div/div[3]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/ul/li[2]/button/div[2]/div[1]"))
            )
            if suggestions:
                ActionChains(driver).move_to_element(suggestions[0]).click().perform()
            else:
                print("No autocomplete suggestions found.")
                driver.quit()
                return
        except Exception as e:
            print("Error handling autocomplete suggestions: ", e)
            driver.quit()
            return

    except Exception as e:
        print("Error interacting with search box: ", e)
        driver.quit()
        return
    
    # Wait for the search results to load
    time.sleep(2)
    
    # Scroll to the bottom to load all hostels (this may need to be adjusted based on the website's behavior)
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        # Scroll down to the bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        # Wait for new page segment to load
        time.sleep(2)  # Adjust the sleep time if needed
        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

#Scraping Hostel Name:

name=[]
a=driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in a:
    b=i.text
    name.append(b)
name=name[2:]


# Scraping Distance from the City:

distance=[]
a=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in a:
    b=i.text
    distance.append(b)    

    
# Scraping Ratings:

rating=[]
a=driver.find_elements(By.XPATH,'//span[@class="score"]')
for i in a:
    b=i.text
    rating.append(b)
rating=rating[2:]

# Scraping Total Review:

comment=[]
a=driver.find_elements(By.XPATH,'//span[@class="keyword"]')
for i in a:
    b=i.text
    comment.append(b)
comment=comment[2:]

# Scraping overall Reiviews:

number=[]
a=driver.find_elements(By.XPATH,'//span[@class="num-reviews left-margin"]')
for i in a:
    b=i.text
    number.append(b)
    
# Scraping Privates from Price:

alle=[]
a=driver.find_elements(By.XPATH,'//div[@class="property-accommodation-price"]')
for i in a:
    b=i.text
    alle.append(b)
alle=alle[2:]
alle=[w.replace('\n',' to ') for w in alle]
first=alle[0::2]

# Scraping Dorms from Price:

second=alle[1::2]

# Scraping Property Description:

allurl=[]                     # Fetching URL's in allurl
start=0
end=1
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//div[@class="property-card"]/a')
    for i in url:
        allurl.append(i.get_attribute('href'))
        
descri=[]                     # Fetching Property Description by opening each URL
for i in allurl:
    driver.get(i)
    time.sleep(5)
    try:
        a=driver.find_element(By.XPATH,'//div[@class="about-description"]/p')
        descri.append(a.text)
    except NoSuchElementException:
        descri.append('Not Present')   
        
descri=descri[2:]    
        
# Checking the length of each element:

print(len(name),len(distance),len(rating),len(comment),len(number),len(first),len(second),len(descri))
```
