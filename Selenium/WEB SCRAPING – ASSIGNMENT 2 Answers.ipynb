Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
salary filter.  
You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results.  
You have to scrape the job-title, job-location, company name, experience required.  
The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs  
The task will be done as shown in the below steps:  
1. first get the web page https://www.naukri.com/ 
2. Enter ‚ÄúData Scientist‚Äù in ‚ÄúSkill, Designations, and Companies‚Äù field.  
3. Then click the search button.  
4. Then apply the location filter and salary filter by checking the respective boxes  
5. Then scrape the data for the first 10 jobs results you get.  
6. Finally create a dataframe of the scraped data.

```python
# Set up the Selenium WebDriver and opening the weblink in automated browser
driver= webdriver.Edge()

driver.get("https://www.naukri.com/")

#to maximize browser window
driver.maximize_window()
```

```python
# Find the search input field and enter the search query
designation = driver.find_element(By.CLASS_NAME,"suggestor-input").send_keys("Data Scientist")
search=driver.find_element(By.CLASS_NAME,"qsbSubmit").click()
```

```python
#Ticking the Location and Salary checkboxes
checkbox_loc=driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[4]/div[2]/div[3]/label/p/span[1]')
checkbox_loc.click()
time.sleep(2)
checkbox_sal=driver.find_element(By.XPATH,"//span[@title='3-6 Lakhs']")
checkbox_sal.click()
```

```python
#Creating empty list to store data
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#Scraping the first 10 Job titles from the given page
titles=driver.find_elements(By.XPATH,'//a[@class="title "]')
for i in titles[0:10]:
    titles=i.text
    job_title.append(titles)

#Scraping the first 10 Job locations from the given page
locations=driver.find_elements(By.CLASS_NAME,"locWdth")
for i in locations[0:10]:
    locations=i.text
    job_location.append(locations)
#Scraping the first 10 Company names from the given page
names=driver.find_elements(By.XPATH,'//a[contains(@class," comp-name ")]')
for i in names[0:10]:
    names=i.text
    company_name.append(names)

#Scraping the first 10 Job experience required details from the given page
exp=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in exp[0:10]:
    exp=i.text
    experience_required.append(exp)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
#Creating the Dataframe DSdf from the Scraped data
DSdf=pd.DataFrame({'Title':job_title,'Location':job_location,'Company Name':company_name, 'Experience':experience_required})
DSdf
```

```python
driver.quit()
```
