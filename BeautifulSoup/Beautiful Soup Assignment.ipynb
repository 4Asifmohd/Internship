# Question 5: Scraping Headlines, Date and News link of CNBC.


```python
page = requests.get('https://www.cnbc.com/world/?region=world')

soup = BeautifulSoup(page.content)

url=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    url.append(i.get('href'))

date=[]
for i in soup.find_all('span',class_="LatestNews-wrapper"):
    date.append(i.text)

headline=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)

print(len(url),len(date),len(headline))

CNBC = pd.DataFrame({'Headings':headline,'Date':date,'News Link':url})
CNBC
```

# Question 6: Scraping most downloaded articels from KeaiPublishing.com

```python
page = requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')

soup = BeautifulSoup(page.content)

title=[]
for i in soup.find_all('h2',class_="h5 article-title"):
    title.append(i.text.replace('\n',''))

date=[]
for i in soup.find_all('p',class_="article-date"):
    date.append(i.text)

author=[]
for i in soup.find_all('p',class_="article-authors"):
    author.append(i.text)

print(len(title),len(date),len(author))

Articles = pd.DataFrame({'Paper Title':title,'Date':date,'Author':author})
Articles
```

1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data 
https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
page = requests.get('https://www.imdb.com/list/ls056092300')

soup = BeautifulSoup(page.content)

name = [] # blank list for adding titles
for i in soup.find_all('h3',class_="ipc-title__text"):
    name.append(i.text)

rating = [] # blank list for adding titles
for i in soup.find_all('span',class_="ipc-rating-star--rating"):
    rating.append(i.text)
   
yearofrelease = [] 
for i in soup.find_all('span',class_="sc-b189961a-8 hCbzGp dli-title-metadata-item"):
    yearofrelease.append(i.text)

print(len(name),len(rating),len(yearofrelease))


IMDB_top_100 = pd.DataFrame({'Name':name,'Rating':rating,'Year of release':yearofrelease})
IMDB_top_100
```
